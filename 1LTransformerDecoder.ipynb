```python
# ============================================
# Imports
# ============================================
import torch
import torch.nn as nn
import torch.nn.functional as F
import math, random, os
from tqdm import trange
import matplotlib.pyplot as plt


device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)


# ============================================
# Dataset: Modular Multiplication
# ============================================

def set_seed(seed):
    random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

def generate_modmul_dataset(p=97, train_frac=0.3, seed=0):
    """
    Input:  [x, y, EQ]
    Label: (x * y) mod p
    """
    set_seed(seed)

    EQ_TOKEN = p  # '=' token
    all_seqs, all_labels = [], []

    for x in range(p):
        for y in range(p):
            seq = torch.tensor([x, y, EQ_TOKEN], dtype=torch.long)
            label = (x * y) % p
            all_seqs.append(seq)
            all_labels.append(label)

    X = torch.stack(all_seqs)
    y = torch.tensor(all_labels)

    n = len(X)
    idx = list(range(n))
    random.shuffle(idx)

    split = int(train_frac * n)
    train_idx = idx[:split]
    test_idx  = idx[split:]

    return X[train_idx], y[train_idx], X[test_idx], y[test_idx]


# ============================================
# Decoder-Only Transformer (GPT-style)
# ============================================

class CausalSelfAttention(nn.Module):
    def __init__(self, d_model, n_heads):
        super().__init__()
        assert d_model % n_heads == 0
        self.d_k = d_model // n_heads
        self.n_heads = n_heads

        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.proj = nn.Linear(d_model, d_model)

    def forward(self, x):
        B, L, D = x.shape
        q = self.W_q(x).view(B, L, self.n_heads, self.d_k).transpose(1, 2)
        k = self.W_k(x).view(B, L, self.n_heads, self.d_k).transpose(1, 2)
        v = self.W_v(x).view(B, L, self.n_heads, self.d_k).transpose(1, 2)

        att = (q @ k.transpose(-1, -2)) / math.sqrt(self.d_k)
        mask = torch.triu(torch.ones(L, L, device=x.device), diagonal=1)
        att = att.masked_fill(mask == 1, float('-inf'))
        att = att.softmax(dim=-1)

        out = att @ v
        out = out.transpose(1, 2).reshape(B, L, D)
        return self.proj(out)

class FeedForward(nn.Module):
    def __init__(self, d_model, d_ff):
        super().__init__()
        self.fc1 = nn.Linear(d_model, d_ff)
        self.fc2 = nn.Linear(d_ff, d_model)

    def forward(self, x):
        return self.fc2(F.relu(self.fc1(x)))

class DecoderBlock(nn.Module):
    def __init__(self, d_model, n_heads, d_ff):
        super().__init__()
        self.ln1 = nn.LayerNorm(d_model)
        self.attn = CausalSelfAttention(d_model, n_heads)
        self.ln2 = nn.LayerNorm(d_model)
        self.ff = FeedForward(d_model, d_ff)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))
        x = x + self.ff(self.ln2(x))
        return x

class TinyGPT(nn.Module):
    def __init__(self, vocab_size, seq_len=3,
                 d_model=128, n_heads=4, d_ff=512, n_layers=2):
        super().__init__()
        self.vocab_size = vocab_size
        self.seq_len = seq_len

        self.token_emb = nn.Embedding(vocab_size, d_model)
        self.pos_emb = nn.Parameter(torch.randn(1, seq_len, d_model))

        self.blocks = nn.ModuleList([
            DecoderBlock(d_model, n_heads, d_ff)
            for _ in range(n_layers)
        ])

        self.ln_f = nn.LayerNorm(d_model)
        self.head = nn.Linear(d_model, vocab_size)

    def forward(self, x):
        B, L = x.shape
        h = self.token_emb(x) + self.pos_emb[:, :L]
        for block in self.blocks:
            h = block(h)
        h = self.ln_f(h)
        logits = self.head(h)
        return logits


# ============================================
# Evaluation Utilities
# ============================================

def evaluate_modmul(model, X_train, y_train, X_test, y_test):
    model.eval()
    with torch.no_grad():
        train_logits = model(X_train)
        test_logits  = model(X_test)

        if train_logits.ndim == 3:
            train_logits = train_logits[:, -1, :]
            test_logits  = test_logits[:, -1, :]

        train_loss = F.cross_entropy(train_logits, y_train).item()
        test_loss  = F.cross_entropy(test_logits,  y_test).item()
        train_acc = (train_logits.argmax(-1) == y_train).float().mean().item()
        test_acc  = (test_logits.argmax(-1)  == y_test).float().mean().item()

    return train_loss, test_loss, train_acc, test_acc


# ============================================
# Training Loop for Decoder-Only Modular Multiplication
# ============================================

def train_decoder_modmul(
    model,
    X_train, y_train,
    X_test, y_test,
    optimizer,
    num_steps=4000,
    log_every=200,
    batch_size=None
):
    steps, train_losses, test_losses = [], [], []
    train_accs, test_accs = [], []

    N = len(X_train)
    use_minibatch = batch_size is not None

    for step in trange(1, num_steps + 1):

        model.train()

        if use_minibatch:
            idx = torch.randint(0, N, (batch_size,), device=X_train.device)
            xb, yb = X_train[idx], y_train[idx]
        else:
            xb, yb = X_train, y_train

        logits = model(xb)
        if logits.ndim == 3:
            logits = logits[:, -1, :]

        loss = F.cross_entropy(logits, yb)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if step % log_every == 0:
            train_loss, test_loss, train_acc, test_acc = evaluate_modmul(
                model, X_train, y_train, X_test, y_test
            )

            steps.append(step)
            train_losses.append(train_loss)
            test_losses.append(test_loss)
            train_accs.append(train_acc)
            test_accs.append(test_acc)

            print(
                f"step {step:5d} | "
                f"train_loss={train_loss:.4f} test_loss={test_loss:.4f} | "
                f"train_acc={train_acc:.3f} test_acc={test_acc:.3f}"
            )

    return steps, train_losses, test_losses, train_accs, test_accs


# ============================================
# Plotting
# ============================================

def plot_training_curves(steps, train_accs, test_accs, train_losses, test_losses):
    plt.figure(figsize=(10, 3))

    plt.subplot(1, 2, 1)
    plt.plot(steps, train_accs, label="train_acc")
    plt.plot(steps, test_accs,  label="test_acc")
    plt.xlabel("step")
    plt.ylabel("accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(steps, train_losses, label="train_loss")
    plt.plot(steps, test_losses, label="test_loss")
    plt.xlabel("step")
    plt.ylabel("loss")
    plt.legend()

    plt.tight_layout()
    plt.show()


# ============================================
# RUN EVERYTHING
# ============================================

PRIME_P = 97
TRAIN_FRAC = 0.3
SEED = 0
SEQ_LEN = 3
VOCAB_SIZE = PRIME_P + 1  # include '=' token

X_train, y_train, X_test, y_test = generate_modmul_dataset(
    p=PRIME_P,
    train_frac=TRAIN_FRAC,
    seed=SEED
)

X_train, y_train = X_train.to(device), y_train.to(device)
X_test,  y_test  = X_test.to(device),  y_test.to(device)

model = TinyGPT(
    vocab_size=VOCAB_SIZE,
    seq_len=SEQ_LEN,
    d_model=64,
    n_heads=2,
    d_ff=128,
    n_layers=1,
).to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1.0)

steps, tr_losses, te_losses, tr_accs, te_accs = train_decoder_modmul(
    model,
    X_train, y_train,
    X_test, y_test,
    optimizer,
    num_steps=20000,
    log_every=200,
    batch_size=2048
)

plot_training_curves(steps, tr_accs, te_accs, tr_losses, te_losses)
```

```python
# ============================================
# Grokking Analysis: Weights + Activations
# ============================================
model.eval()

# -----------------------------
# 1. Fourier analysis of weights
# -----------------------------

# Token embeddings: [vocab_size, d_model]
tok_emb = model.token_emb.weight.detach().cpu()          # [V, D]

# Output head: either [V, D] or [D, V]
head_w = model.head.weight.detach().cpu()
if head_w.shape[0] == tok_emb.shape[1]:
    # If it's [D, V], transpose to [V, D]
    head_w = head_w.T                                    # [V, D]

def fourier_spectrum_over_tokens(W: torch.Tensor) -> torch.Tensor:
    """
    Compute average Fourier magnitude over embedding dims,
    treating the token index as the 'time' axis.
    W: [V, D]
    Returns: [V] magnitude per frequency index.
    """
    fft = torch.fft.fft(W, dim=0)   # [V, D]
    mag = fft.abs()
    avg_mag = mag.mean(dim=1)       # [V]
    return avg_mag

tok_spec  = fourier_spectrum_over_tokens(tok_emb)
head_spec = fourier_spectrum_over_tokens(head_w)

fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(tok_spec.numpy())
axs[0].set_title("Token Embedding Fourier Spectrum")
axs[0].set_xlabel("Frequency index k")
axs[0].set_ylabel("Avg magnitude")

axs[1].plot(head_spec.numpy())
axs[1].set_title("Head Weight Fourier Spectrum")
axs[1].set_xlabel("Frequency index k")
axs[1].set_ylabel("Avg magnitude")

plt.tight_layout()
plt.show()

print("Top 5 frequencies (token embeddings):")
tok_top = torch.topk(tok_spec, k=5)
for mag, idx in zip(tok_top.values, tok_top.indices):
    print(f"  k={int(idx):3d}, magnitude={float(mag):.4f}")

print("\nTop 5 frequencies (head weights):")
head_top = torch.topk(head_spec, k=5)
for mag, idx in zip(head_top.values, head_top.indices):
    print(f"  k={int(idx):3d}, magnitude={float(mag):.4f}")

# ------------------------------------------------
# 2. Activation analysis on all (x, y) mod PRIME_P
# ------------------------------------------------

# Build the full dataset of all pairs (x, y) with '=' token
p = PRIME_P  # from earlier in the notebook
EQ_TOKEN = p

all_seqs = []
all_labels = []
for x in range(p):
    for y in range(p):
        all_seqs.append([x, y, EQ_TOKEN])
        all_labels.append((x * y) % p)

X_all = torch.tensor(all_seqs, dtype=torch.long, device=device)  # [N, 3]
y_all = torch.tensor(all_labels, dtype=torch.long, device=device)# [N]
N = X_all.shape[0]

print(f"\nBuilt full dataset of size N = {N} = {p} * {p}")

# Hook into the last decoder block to capture hidden activations
activations = {}

def block_hook(module, inp, out):
    # out: [N, seq_len, d_model]
    activations["block_out"] = out.detach().cpu()

# Register hook on the last block
hook_handle = model.blocks[-1].register_forward_hook(block_hook)

with torch.no_grad():
    _ = model(X_all)

hook_handle.remove()

hidden = activations["block_out"]       # [N, seq_len, d_model]
hidden_last = hidden[:, -1, :]          # take last token position: [N, d_model]
H = hidden_last.numpy()
N, d_model = H.shape

print(f"Captured hidden states: shape {hidden.shape}, using last position: {hidden_last.shape}")

# -------------------------------------
# 3. Correlate neurons with math features
# -------------------------------------

x_vals = X_all[:, 0].cpu().numpy()
y_vals = X_all[:, 1].cpu().numpy()

features = {
    "x": x_vals,
    "y": y_vals,
    "x_plus_y_mod_p": (x_vals + y_vals) % p,
    "x_minus_y_mod_p": (x_vals - y_vals) % p,
}

def corr(a, b):
    a = a.astype(float) - a.mean()
    b = b.astype(float) - b.mean()
    denom = (np.linalg.norm(a) * np.linalg.norm(b))
    if denom == 0:
        return 0.0
    return float((a * b).sum() / denom)

results = {}

for name, f in features.items():
    corrs = []
    for j in range(d_model):
        c = corr(H[:, j], f)
        corrs.append(c)
    corrs = np.array(corrs)
    results[name] = corrs

    j_max = int(np.argmax(np.abs(corrs)))
    print(f"\nFeature '{name}':")
    print(f"  best neuron index = {j_max}, corr = {corrs[j_max]:.4f}")

    # Optional: plot sorted absolute correlations for this feature
    sorted_abs = np.sort(np.abs(corrs))[::-1]
    plt.figure(figsize=(5, 3))
    plt.plot(sorted_abs)
    plt.title(f"Neuron |corr| sorted for feature '{name}'")
    plt.xlabel("Neuron (sorted)")
    plt.ylabel("|corr|")
    plt.tight_layout()
    plt.show()
```

```python
# ============================================
# Fourier Components of Embedding Matrix
# ============================================

model.eval()

# Get embedding matrix [V, D]
W = model.token_emb.weight.detach().cpu().numpy()
V, D = W.shape

# Exclude "=" token (optional, matches paper more closely)
p = PRIME_P
W = W[:p]       # Now W is [p, D]

# Token values x = 0..p-1
xs = np.arange(p)

# Max frequency = p // 2 (Nyquist)
K = p // 2

sin_norms = []
cos_norms = []

for k in range(K):
    # sin(2πkx/p) and cos(2πkx/p) as column vectors [p, 1]
    sin_wave = np.sin(2 * np.pi * k * xs / p)[:, None]   # [p,1]
    cos_wave = np.cos(2 * np.pi * k * xs / p)[:, None]   # [p,1]

    # Project embedding onto these waves
    # Equivalent to W^T sin_wave (summed over tokens)
    proj_sin = (W * sin_wave).sum(axis=0)   # [D]
    proj_cos = (W * cos_wave).sum(axis=0)   # [D]

    # Store norms (L2)
    sin_norms.append(np.linalg.norm(proj_sin))
    cos_norms.append(np.linalg.norm(proj_cos))

# Convert to arrays
sin_norms = np.array(sin_norms)
cos_norms = np.array(cos_norms)

# -------------------------------
# Plot in research-paper style
# -------------------------------
plt.figure(figsize=(8, 4))

plt.stem(range(K), sin_norms, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.stem(range(K), cos_norms, linefmt='r-', markerfmt='ro', basefmt=' ')

plt.title("Fourier Components of Embedding Matrix", fontsize=14)
plt.xlabel("Frequency k")
plt.ylabel("Norm of Fourier Component")

plt.legend(["sin", "cos"])
plt.tight_layout()
plt.show()

```

```python
# ============================================
# Activation Analysis: FFT + PCA of Hidden States
# ============================================
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

model.eval()
device = next(model.parameters()).device

p = PRIME_P          # prime modulus
EQ_TOKEN = p         # "=" token index

# -------------------------------------------------
# 1. Build full dataset of all (x, y, "=") pairs
# -------------------------------------------------
all_seqs = []
all_x = []
all_y = []

for x in range(p):
    for y in range(p):
        all_seqs.append([x, y, EQ_TOKEN])
        all_x.append(x)
        all_y.append(y)

X_all = torch.tensor(all_seqs, dtype=torch.long, device=device)  # [p^2, 3]
N = X_all.shape[0]
all_x = np.array(all_x)
all_y = np.array(all_y)

print(f"[Activation Analysis] Using N = {N} = {p} * {p} sequences")

# -------------------------------------------------
# 2. Capture last-layer, last-token activations
# -------------------------------------------------
activations = {}

def hook_last_block(module, inp, out):
    # out: [N, seq_len, d_model]
    activations["last_hidden"] = out.detach().cpu()

hook_handle = model.blocks[-1].register_forward_hook(hook_last_block)

with torch.no_grad():
    _ = model(X_all)

hook_handle.remove()

hidden = activations["last_hidden"]              # [N, seq_len, d_model]
hidden_last = hidden[:, -1, :]                   # [N, d_model]
N, d_model = hidden_last.shape
print(f"[Activation Analysis] Hidden last shape: {hidden_last.shape}")

# Reshape to [p, p, d_model] so we can treat x and y separately
hidden_grid = hidden_last.numpy().reshape(p, p, d_model)   # [x, y, d]

# -------------------------------------------------
# 3. Fourier analysis over x (FFT of activations)
# -------------------------------------------------
# FFT along x axis (axis=0), then take magnitude
fft_x = np.fft.fft(hidden_grid, axis=0)          # [p, p, d]
mag_x = np.abs(fft_x)

# Energy per (k, neuron) averaged over y
energy_x = (mag_x ** 2).mean(axis=1)             # [p, d]

# Average over neurons to get a global spectrum vs frequency k
avg_energy_per_k = energy_x.mean(axis=1)         # [p]

plt.figure(figsize=(10, 4))
plt.plot(avg_energy_per_k, marker='o', linewidth=1)
plt.title("Avg Fourier Energy of Last-Layer Activations (over neurons & y)")
plt.xlabel("Frequency k (over x)")
plt.ylabel("Energy (mean |FFT|^2)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Optional: print top few frequencies by energy (excluding k=0)
k_sorted = np.argsort(-avg_energy_per_k)
print("\nTop 5 non-zero frequencies by activation energy (over x):")
count = 0
for k in k_sorted:
    if k == 0:
        continue
    print(f"  k = {int(k):3d}, energy = {avg_energy_per_k[k]:.6e}")
    count += 1
    if count >= 5:
        break

# -------------------------------------------------
# 4. PCA of activations (geometry in activation space)
# -------------------------------------------------
A = hidden_last.numpy()     # [N, d_model]

# 2D PCA
pca_2d = PCA(n_components=2)
A_2d = pca_2d.fit_transform(A)

print(f"\n[PCA] Explained variance (PC1, PC2): {pca_2d.explained_variance_ratio_}")

plt.figure(figsize=(12, 5))

# Left: colored by x
plt.subplot(1, 2, 1)
sc1 = plt.scatter(A_2d[:, 0], A_2d[:, 1], c=all_x, s=5, cmap="rainbow")
plt.colorbar(sc1, label="Input x")
plt.title("PCA of Hidden States (Colored by x)")
plt.xlabel("PC 1")
plt.ylabel("PC 2")

# Right: colored by y
plt.subplot(1, 2, 2)
sc2 = plt.scatter(A_2d[:, 0], A_2d[:, 1], c=all_y, s=5, cmap="rainbow")
plt.colorbar(sc2, label="Input y")
plt.title("PCA of Hidden States (Colored by y)")
plt.xlabel("PC 1")
plt.ylabel("PC 2")

plt.tight_layout()
plt.show()

print("\n[Activation Analysis] Done: FFT spectrum + PCA of hidden states.")
```
